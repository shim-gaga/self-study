### ì§€ëŠ¥í˜• ë¦¬ìŠ¤íŠ¸ (Comprehending Lists)
ë°ì´í„°ì˜ ì–‘ì´ í˜„ì €íˆ ë§ì•„ì§ˆ ê²½ìš° list comprehensionì˜ ì†ë„ê°€ ë” ìš°ìˆ˜í•˜ë‹¤.

```python
# non comprehending lists
chars = '!@#%$^&*()'
code1 = []

for s in chars:
    code1.append(ord(s))
    
print(code1)
```

```python
# comprehending lists
code2 = [ord(s) for s in chars]

print(code2)
```

```python
# comprehending lists
code3 = [ord(s) for s in chars if ord(s) > 40]

print(code3)
```

### List Comprehension + Map, Filter

```python
code4 = list(filter(lambda x: x > 40, map(ord, chars)))

print(code4)
```

```python
print([chr(s) for s in code1])
print([chr(s) for s in code2])
print([chr(s) for s in code3])
print([chr(s) for s in code4])
```

### Dictionaryì˜ ë‚´ë¶€ êµ¬ì¡°

```python
# ì‚¬ìš©ì ì •ì˜ dict ìƒì†(UserDict ê°€ëŠ¥)
# ì‚¬ìš©ì ì •ì˜ í´ë˜ìŠ¤ë¥¼ ë§Œë“¤ì–´ì„œ dict ì‚¬ìš©
# ë”•ì…”ë„ˆë¦¬ì˜ ë‚´ë¶€ êµ¬ì¡°: dictionaryë¥¼ ì„ ì–¸í•˜ë©´ ì´ë ‡ê²Œ ì‘ë™í•œë‹¤

class UserDict(dict):
    def __missing__(self, key):
        print('Called: __missing__')
        if isinstance(key, str):
            raise KeyError(key)
        return self[str(key)]
    
    def get(self, key, default=None):
        print("Called: __getitem__")
        try:
            return self[key]
        except KeyError:
            return default
    
    def __contains__(self, key):
        print("Called: __contains__")
        return key in self.keys() or str(key) in self.keys()
```

```python
user_dict1 = UserDict(one=1, two=2)
user_dict2 = UserDict({'one':1, 'two':2})
user_dict3 = UserDict([('one', 1), ('two', 2)])

# ì¶œë ¥
print(user_dict1)
print(user_dict2)
print(user_dict3)
```
