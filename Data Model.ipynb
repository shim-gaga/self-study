ë„¤ì„ë“œ íŠœí”Œì—ì„œ ì•„ë˜ì˜ 3ê°€ì§€ë¥¼ ê¸°ì–µí•´ì•¼ í•œë‹¤.

```python
# ì¼ë°˜ì ì¸ íŠœí”Œ ì‚¬ìš©
# ë‘ ì  ì‚¬ì´ì˜ ê±°ë¦¬: [0] - x, [1] - y. ì´ë ‡ê²Œ ë ˆì´ë¸”ì„ ë‹¬ì•„ì¤˜ì•¼ í•œë‹¤.
pt1 = (1.0, 5.0)
pt2 = (2.5, 1.5)

from math import sqrt

line_leng1 = sqrt((pt2[0] - pt1[0]) ** 2 + (pt2[1] - pt1[1]) ** 2)
print(line_leng1)
```

```python
# ë„¤ì„ë“œ íŠœí”Œ ì‚¬ìš©: íŠœí”Œì˜ ì„±ì§ˆë„ ê°€ì§€ê³  ìˆê³  ë§ˆì¹˜ í´ë˜ìŠ¤ë¥¼ ê°€ì§€ê³  ìˆëŠ” ê²ƒ ê°™ìŒ
from collections import namedtuple

# ë„¤ì„ë“œ íŠœí”Œ ì„ ì–¸
Point = namedtuple('Point', 'x y')
pt1 = Point(1.0, 5.0)
pt2 = Point(2.5, 1.5)

line_leng2  = sqrt((pt2.x - pt1.x) ** 2 + (pt2.y - pt1.y) ** 2)
print(line_leng2)
```

```python
print(line_leng1 == line_leng2)
```

### ë„¤ì„ë“œ íŠœí”Œ ì„ ì–¸ ë°©ë²•

```python
print(Point1, Point2, Point3, Point4)
```

```python
# ì‹¤ ì‚¬ìš© ì‹¤ìŠµ
# í•™ìƒ ì „ì²´ ê·¸ë£¹ ìƒì„±
# ë°˜ 20ëª…, 4ê°œì˜ ë°˜ -> (A, B, C, D) ë²ˆí˜¸

# ë„¤ì„ë“œ íŠœí”Œ ì„ ì–¸. ë³€ìˆ˜ì™€ ë„¤ì„ë“œ íŠœí”Œ ëª…ì„ ê°™ê²Œ í•˜ëŠ” ê²ƒì´ ê´€ë¡€
Classes = namedtuple('Classes', ['rank', 'number'])

# ê·¸ë£¹ ë¦¬ìŠ¤íŠ¸ ì„ ì–¸
numbers = [str(n) for n in range(1, 21)]
ranks = 'A B C D'.split()
```

```python
# List Comprehension
students = [Classes(rank, number) for rank in ranks for number in numbers]
print(students)
```

```python
# ê°€ë…ì„±ì´ ì•ˆì¢‹ì€ ì¼€ì´ìŠ¤
students2 = [Classes(rank, number) for rank in 'A B C D'.split()
            for number in [str(n) for n in range(1, 21)]]
print(students2)
```

```python
# ì¶œë ¥
for s in students:
    print(s)
```

```python
# í´ë˜ìŠ¤ ì˜ˆì œ
class Student:
    def __init__(self, name, height):
        self._name = name
        self._height = height
        
    def __str__(self):
        return "Student Class Info : {}, {}".format(self._name, self._height)
    
    def __ge__(self, x):
        print("Called >> __ge__ Method.")
        
        if self._height >= x._height:
            return True
        else:
            return False
        
    def __le__(self, x):
        print("Called >> __le__ Method.")
        
        if self._height <= x._height:
            return True
        else:
            return False
        
    def __sub__(self, x):
        print("Called >> __sub__ Method.")
        return self._height - x._height
```

```python
# ì¸ìŠ¤í„´ìŠ¤ ìƒì„±
s1 = Student("James", 181)
s2 = Student("Mie", 165)

print(s1 + s2) # ë§¤ì§ ë©”ì†Œë“œë¥¼ ì˜¤ë²„ë¼ì´ë”© í•˜ì§€ ì•Šì•˜ìœ¼ë¯€ë¡œ ê³„ì‚° ë¶ˆê°€
```

```python
# ë§¤ì§ ë©”ì†Œë“œ ì¶œë ¥
print(s1 >= s2)
print(s1 <= s2)
print(s1 - s2)
print(s2 - s1)
```

```python
# í´ë˜ìŠ¤ ì˜ˆì œ 2
# ë²¡í„°
class Vector(object):
    def __init__(self, *args):
        '''Create a vector, example : v = Vector(1,2)'''
        if len(args) == 0:
            self._x, self._y = 0, 0
        else:
            self._x, self._y = args
            
    def __repr__(self):
        '''Returns the vector informations'''
        # %s ì§€ì •ìëŠ” str() ì„ ì‚¬ìš©í•˜ì—¬ ê°ì²´ë¥¼ ë³€í™˜í•˜ê³  %rëŠ” repr() ì„ ì‚¬ìš©í•˜ì—¬ ê°ì²´ë¥¼ ë³€í™˜
        return 'Vector(%r, %r)' % (self._x, self._y)
    
    def __add__(self, other):
        '''Returns the vector addition of self and other'''
        return Vector(self._x + other._x, self._y + other._y)
    
    
    def __mul__(self, y):
        '''Returns the vector multiplication of self and other'''
        return Vector(self._x * y, self._y * y)
    
    def __bool_(self):
        return bool(max(self._x, self._y))
```

```python
# Vecotr ì¸ìŠ¤í„´ìŠ¤ ìƒì„±
v1 = Vector(3, 5)
v2 = Vector(15, 20)
v3 = Vector()
```

```python
print(Vector.__init__.__doc__)
print(Vector.__repr__.__doc__)
print(Vector.__add__.__doc__)
```

```python
print(v1, v2, v3)
```

```python
print(v1 + v2)
print(v1 * 4)
print(bool(v1), bool(v2))
print(bool(v3))
```
