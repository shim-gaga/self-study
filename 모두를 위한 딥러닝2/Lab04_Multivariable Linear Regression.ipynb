## Hypothesis Function


ÏûÖÎ†• Î≥ÄÏàòÍ∞Ä 3Í∞úÎùºÎ©¥ wieghtÎèÑ 3Í∞úÏùº Í≤ÉÏù¥Îã§. Ïù¥Îïå, xÏùò Îç∞Ïù¥ÌÑ∞Ïùò Í∏∏Ïù¥Í∞Ä Îß§Ïö∞ Í∏∏ ÏàòÎèÑ ÏûàÍ∏∞ ÎïåÎ¨∏Ïóê matmul()Î°ú ÌïúÎ≤àÏóê Í≥ÑÏÇ∞ÌïúÎã§. Ïù¥ Î∞©Î≤ïÏù¥ Îçî Í∞ÑÍ≤∞ÌïòÍ≥†, xÏùò Í∏∏Ïù¥Í∞Ä Î∞îÎÄåÏñ¥ÎèÑ ÏΩîÎìúÎ•º Î∞îÍøÄ ÌïÑÏöîÍ∞Ä ÏóÜÏúºÎ©∞, ÏÜçÎèÑÎèÑ Îçî Îπ†Î•¥Îã§

## Cost function
### MSE
ÎπÑÏö© Ìï®ÏàòÎäî Í∏∞Ï°¥ Simple Linear RegressionÍ≥º ÎèôÏùºÌïú Í≥µÏãùÏùÑ ÏÇ¨Ïö©ÌïúÎã§.  


### Gradient Descent with torch.optim


### Full Code with torch.optim

```python
import torch
```

```python
# Îç∞Ïù¥ÌÑ∞
x_train = torch.FloatTensor([[73, 80, 75],
                             [93, 88, 93],
                             [89, 91, 90],
                             [96, 98, 100],
                             [73, 66, 70]])
y_train = torch.FloatTensor([[152], [185], [180], [196], [142]])

# Î™®Îç∏ Ï¥àÍ∏∞Ìôî
W = torch.zeros((3, 1), requires_grad=True)
b = torch.zeros(1, requires_grad=True)

# optimizer ÏÑ§Ï†ï
optimizer = torch.optim.SGD([W, b], lr=1e-5)
```

```python
nb_epochs = 1000

for epoch in range(1, nb_epochs + 1):
    hypothesis = x_train.matmul(W) + b
    cost = torch.mean((hypothesis - y_train) ** 2)
    
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    
    print('Epoch {:4d}/{} hypothesis: {} Cost: {:.6f}'.format(
        epoch, nb_epochs, hypothesis.squeeze().detach(), cost.item()))
   
```

### Custom Dataset
torch.utils.data.DatasetÏùÑ ÏÉÅÏÜçÎ∞õÏïÑ ÏßÅÏ†ë Ïª§Ïä§ÌÖÄ Îç∞Ïù¥ÌÑ∞ÏÖã(Custom Dataset)ÏùÑ ÎßåÎìúÎäî Í≤ΩÏö∞ÎèÑ ÏûàÎã§.   
torch.utils.data.DatasetÏùÄ ÌååÏù¥ÌÜ†ÏπòÏóêÏÑú Îç∞Ïù¥ÌÑ∞ÏÖãÏùÑ Ï†úÍ≥µÌïòÎäî Ï∂îÏÉÅ ÌÅ¥ÎûòÏä§Ïù¥Îã§.  
  
Ïª§Ïä§ÌÖÄ Îç∞Ïù¥ÌÑ∞ÏÖãÏùÑ ÎßåÎì§ Îïå, Í∏∞Î≥∏Ï†ÅÏù∏ ÎºàÎåÄÎäî ÏïÑÎûòÏôÄ Í∞ôÎã§.  
Ïó¨Í∏∞ÏÑú ÌïÑÏöîÌïú Í∏∞Î≥∏Ï†ÅÏù∏ Î©îÏÜåÎìúÎäî \_\_init\_\_(), \_\_len\_\_(), \_\_getiem\_\_() Ïù¥Îã§.

```python
# ÏÉàÎ°úÏö¥ custom ÌÅ¥ÎûòÏä§Î•º ÎßåÎì¶ÏúºÎ°úÏç® ÏõêÌïòÎäî Îç∞Ïù¥ÌÑ∞ ÏÖãÏùÑ ÎßåÎì§ Ïàò ÏûàÎã§
from torch.utils.data import Dataset

class CustomDataset(Dataset):
    # Îç∞Ïù¥ÌÑ∞ÏÖãÏùò Ï†ÑÏ≤òÎ¶¨Î•º Ìï¥Ï£ºÎäî Î©îÏÜåÎìú
    def __init__(self): 
        self.x_data = [[73, 80, 75],
                       [93, 88, 93],
                       [89, 91, 90],
                       [96, 98, 100],
                       [73, 66, 70]]
        self.y_data = [[152], [185], [180], [196], [142]]
            
    # Îç∞Ïù¥ÌÑ∞ÏÖãÏùò Í∏∏Ïù¥, Ï¶â Ï¥ù sampleÏùò ÏàòÎ•º Î¶¨ÌÑ¥ÌïòÎäî Î©îÏÜåÎìú
    def __len__(self): 
        return len(self.x_data)

    # Îç∞Ïù¥ÌÑ∞ÏÖãÏóêÏÑú ÌäπÏ†ï 1Í∞úÏùò ÏÉòÌîåÏùÑ Í∞ÄÏ†∏Ïò§Îäî Î©îÏÜåÎìú
    def __getitem__(self, idx):
        x = torch.FloatTensor(self.x_data[idx])
        y = torch.FloatTensor(self.y_data[idx])
        
        return x, y
```

```python
dataset = CustomDataset()
```

### TensorDataset
TensorDatasetÏùÄ Í∏∞Î≥∏Ï†ÅÏúºÎ°ú ÌÖêÏÑúÎ•º ÏûÖÎ†•ÏúºÎ°ú Î∞õÎäîÎã§. ÌÖêÏÑú ÌòïÌÉúÎ°ú Îç∞Ïù¥ÌÑ∞Î•º Ï†ïÏùòÌïúÎã§.

```python
from torch.utils.data import TensorDataset 

x_train  =  torch.FloatTensor([[73,  80,  75], 
                               [93,  88,  93], 
                               [89,  91,  90], 
                               [96,  98,  100],   
                               [73,  66,  70]])  
y_train  =  torch.FloatTensor([[152],  [185],  [180],  [196],  [142]])
```

```python
from torch.utils.data import DataLoader

dataloader = DataLoader(
    dataset,
    batch_size=2,
    shuffle=True
)
```
